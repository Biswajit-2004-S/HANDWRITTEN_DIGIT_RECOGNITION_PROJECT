### Theory
**The data i.e the handwritten digits to be recognized (are in form of image) , as python does not undestand images thus needs
to be converted into actual numbers(digits) of a 3x3 matrix of array by encoding the data.**

**Thus the mnist dataset already has some preconverted dataset of numbers in a 3x3 array format, we are using this to train our model.**

```python
x_train.shape
```

```python
x_test.shape
```

**Thus we have 60000 images for our training and 10000 images for testing of length x width = 28x28**

**CNN will accept the data in 28 x 28 x 3 format , where 3 is RGB every image consists of red,blue,green channels.**

```python
width,height = 28,28
input_shape = (width,height,1)
```

```python
input_shape
```

### Now re schedule training data 

```python
x_train = x_train.reshape(x_train.shape[0],height,width,1)
x_test = x_test.reshape(x_test.shape[0],height,width,1)
```

```python
x_train.shape
```

```python
x_test.shape
```

**So we have now re-scheduled...**
**Now we are using 1 here insted of 3 as 1 is used for gray scaling i.e black and white images as no red,green,blue channels are present in the dataset**

```python
y_train.shape # It will not change as it has labels here only i.e the single numbers
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size =0.1 )
```

```python
len(x_val)
```

**In black and white images the range is between 0-255 so some images might have a value of 0(black) and some have value of 255(white) so
we need to Normalize our data by Feature Scaling so that our model will focus equally for all the images.**

```python
## Normalizing Data
x_train = (x_train - x_train.mean()) / x_train.std()
x_val = (x_val - x_val.mean()) / x_val.std()
x_test = (x_test - x_test.mean()) / x_test.std()
```

**We have Normalized our training data now we need to covert our testing data too...**

```python
## Initial y_tarin ...
y_train
```

```python
from tensorflow import keras
num_labels=10
y_train=keras.utils.to_categorical(y_train)
```

```python
# The modified data is now...
y_train
```

```python
## Now for validation and testing data also ...
y_val=keras.utils.to_categorical(y_val)
y_test=keras.utils.to_categorical(y_test)
```

## Building Our Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout ,Flatten, Input, Conv2D, MaxPooling2D, AveragePooling2D
# Search for LeNet 5 Architecture to understand more
```

```python
# Creating/Copying the architecture of LeNet 5 
model = Sequential()
model.add(Conv2D(6, kernel_size = (5, 5), strides = (1,1), activation = 'tanh', input_shape = (28,28,1)))
model.add(AveragePooling2D(pool_size = (2,2) , strides = (1,1)))
model.add(Conv2D(16, kernel_size = (5, 5), strides = (1,1), activation = 'tanh'))
model.add(AveragePooling2D(pool_size = (2,2) , strides = (2,2)))
model.add(Conv2D(120, kernel_size = (5, 5), strides = (1,1), activation = 'tanh'))
## This is the feature Extractor Block that we havw done above 

model.add(Flatten())
model.add(Dense(84, activation = 'tanh'))
model.add(Dense(num_labels, activation = 'softmax'))
```

**Let's see our model Layers Summary ...**

```python
model.summary()
```

**Now compiling our model ...**

```python
model.compile(loss = keras.losses.categorical_crossentropy, optimizer = 'adam' , metrics = ['accuracy'])
```

```python
## Training Our model now ...
history = model.fit(x_train, y_train, batch_size = 32, epochs = 20, verbose = 1, validation_data = (x_val, y_val))
```

### So Accuracy we got is 0.9889 & validation_accuracy we got is 0.0575

**In Neural Network the main objective here is to minimize the losses,
thus We are using loss function here which basically
take two values as input the y_pred value and the actual value and find based on that. ---->
We used 'categorical_crossentropy' as we are using multi class classification here. ----> Optimizer is used to fasten our process,
"Adam" is the fastest Optimizer. ---->in history we are training our model by providing our training data and labels here.**

**We are traing for 60000 datas which are being stored in gpu ram memory during execution but gpu ram is limited we cannot 
provide all at a time so we divide it into batches by using 'batch_size' function , here we are dividing the batch into 32 parts
, i.e 32 images will be prcessed at once .**

### Testing our data 

```python
score = model.evaluate(x_test, y_test, verbose = 1)
```
